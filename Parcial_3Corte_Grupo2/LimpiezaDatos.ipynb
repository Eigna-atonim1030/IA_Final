Importanción de librerias

```python
import pandas as pd
import numpy as np
from tensorflow import keras
import matplotlib.pyplot as plt

# Red neuronal secuencial
from tensorflow.keras.models import Sequential
# Capa completamente conectada
from tensorflow.keras import layers
from tensorflow.keras.layers import Dense, Conv2D
# Optimizador
from tensorflow.keras.optimizers import Adam
```

```python
# Cargar los archivos CSV
df1 = pd.read_csv('Estudiantes 2023-2_Listado_PermaneSer - BD.csv')  # fuente de datos más completa
df2 = pd.read_csv('Estado en SIMCA - Sheet0.csv')  # este es el que se quiere completar
```

```python
# Asegúrate que CODIGO sea string y sin espacios para evitar problemas
df1['CODIGO'] = df1['CODIGO'].astype(str).str.strip()
df2['CODIGO'] = df2['CODIGO'].astype(str).str.strip()

# Merge para traer columnas de df1 a df2
df_merged = pd.merge(df2, df1 [['CODIGO', 'CORREO', 'IDENTIFICACION', 'GENERO', 'ESTUDIANTE']], 
                     on='CODIGO', how='left', suffixes=('', '_df1'))

# Actualizar columnas de df2 con las de df1 solo si no están vacías en df1
for col in ['CORREO', 'IDENTIFICACION', 'GENERO', 'ESTUDIANTE']:
    # Si el valor en df1 no es nulo, reemplaza en df_merged
    df_merged[col] = df_merged[col+'_df1'].combine_first(df_merged[col])
    df_merged.drop(columns=[col+'_df1'], inplace=True)

# df_merged ahora tiene todas las filas y columnas actualizadas
df_merged.to_csv('df_completado.csv', index=False)

```
